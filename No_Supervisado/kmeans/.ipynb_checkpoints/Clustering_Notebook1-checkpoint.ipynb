### Notebook 1:  Clustering con K-means
En este notebook vamos a implementar el algoritmo de clustering K-means en R. El objetivo es entender como funciona este algoritmo y  aprender a utilizarlo con un ejemplo simple. Se incluye además la estrategia para definir el número de grupos k. Al final se propone un ejercicio con nuevos datos, con el fin de aprender a utilizar el algoritmo. 

```r
library(cluster)
library(ggplot2)

#ruta datos
#lee datos del archivo de texto como una matriz. La primera #columna se utiliza como vector de nombres de las filas
datos <-read.table("./milk1.dat",row.names=1,dec=".",header=TRUE)
summary(datos)
boxplot(datos)
```

```r
print(datos)
```


**Ejercicio 1:** describa los datos y discuta la pertinencia de realizar un análisis de clustering o agrupamiento

RESPUESTA: el dataset contiene mediciones realizadas a la composicion de muestras de leche de diferentes animales  a la cual se le controlaron los siguientes parametros: agua, proteina, grasa, lactosa y ASH. Es importante destacar que se encuentran en porcentajes por lo que podriamos decir que todos estan en la misma "unidad o escala". podria ser interesante ver si es posible agrupar estos animales en alguna clasificacion realizada en los libros a partir de la clusterizacion de la leche que producen

```r
#ingresar valor de k0 
k0=3
cl<-kmeans(datos,k0,iter.max=1000,nstart=10)
#grafica los datos coloreados con el grupo asignado en planos de pares de variables
pairs(datos,col=cl$cluster)
```

**Ejercicio 2:** Interprete y comente los resultados obtenidos. 

**Repuesta**// me gusta el par que se da en FAT-WATER que permite definir casi en su totalidad los clusters agrupados por kmeans (por lo menos para el ejemplo de 3 clusters), el cluster formado por los animales marinos ( foca y delfin) tiene sentido naturalmente hablando, ya que los dos estan en el agua, por ese lado hace sentido, pero continuando con la observacion se puede apreciar que la ballena quedo en otro grupo con las ratas, zebras y perros, por lo que no tiene mucho sentido esa clusterizacion a simple vista

```r
#Análisis de Componentes Principales para graficar los grupos obtenidos
#Calcula plano de componentes principales y grafica las observaciones con el número de su grupo
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,cl$centers)
res2
cl
biplot(res,xlabs=cl$cluster)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:k0,pch=19)
```

```r
#con K=2
k0=2
cl<-kmeans(datos,k0,iter.max=1000,nstart=10)
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,cl$centers)
res2
cl
biplot(res,xlabs=cl$cluster)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:k0,pch=19)
```

```r
#con K=4
k0=4
cl<-kmeans(datos,k0,iter.max=1000,nstart=10)
res <- princomp(datos, cor=TRUE)
res2 <- predict(res,cl$centers)
res2
cl
biplot(res,xlabs=cl$cluster)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
points(res2[,1:2],col=1:k0,pch=19)
```

```r
#Estrategia para elegir K, el numero de grupos
#J suma cuadrado de las distancias por cada k utilizado
#K-means
J<-rep(0,20)
fin_iter<-20


for (k in 1:fin_iter){
     cl<-kmeans(datos,k,iter.max=1000,nstart=10)
	 J[k]<-cl$tot.withinss
}


plot(1:fin_iter,J,xlab="k",ylab="J",type="p",main="J v/s k en k-means",col=2) 
```

```r
#definir el umbral para encontrar el codo, el umbral indica que porcentaje de variacion el en ratio puede tolerar 
#el algoritmo para definir un k, en este caso si la variacion es superior a un valor de 0.05 se corta y se encuentra
#el k, osea debe mantener un decaimiento constante que no varie en mas de un 5 %

umbral<-0.05



for (k in 1:fin_iter){
     cl<-kmeans(datos,k,iter.max=1000,nstart=10)
	 J[k]<-cl$tot.withinss
    #print(J[k])
    if (k>2){
        ratio= (J[k-1]-J[k])/J[k-1]
        ratio_anterior= (J[k-2]-J[k-1])/J[k-2]
        #print(ratio-ratio_anterior)
        if((ratio_anterior-ratio)<umbral){
            print((J[k-1]-J[k])/J[k-1])
            ratio<- (J[k-1]-J[k])/J[k-1]
            
            
        }
        else{
            print("el k indicado es")
            print(k-1)
            break()
            
        }
    }    
}
```
